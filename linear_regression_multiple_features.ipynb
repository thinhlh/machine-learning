Import libraries

```python
import pandas as pd
```

Read data

```python
data= pd.read_csv('/content/50_Startups.csv')
data.head()
```

Retreive features and predictions values

```python
x= data.iloc[:,:4] # labels
y= data.iloc[:,4] # prediction
```

# Using one hot encoder for state column
## This will convert state from enum type to computer-readable values

```python
x= pd.get_dummies(x,drop_first=True)
x
```

# Splitting data to training and testing set
## Train / Test = 8/2

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test= train_test_split(x,y,test_size=0.2)
```

# Training data using Linear Regression

```python
from sklearn.linear_model import LinearRegression

linearRegression=LinearRegression()
```

Fitting model

```python
linearRegression.fit(x_train,y_train)
```

# Testing model

```python
y_prediction= linearRegression.predict(x_test)
y_prediction
```

Calculating accuracy using r2_score

```python
from sklearn.metrics import r2_score

score=r2_score(y_test,y_prediction)

print(f'Model accuracy: {score*100}%')
```
