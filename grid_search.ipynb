Using grid search to find super parameter

# Import libraries

```python
import numpy as np
import pandas as pd
from sklearn.metrics import classification_report
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split 
from sklearn.model_selection import GridSearchCV
```

Read data

```python
data = pd.read_csv('/content/Social_Network_Ads.csv')
data.describe()
```

Retrieve features and target

```python
features, target = data.drop(['Purchased'], axis=1), data['Purchased']
```

```python
features.describe()
```

```python
target.describe()
```

Split training and testing set

```python
x_train, x_test, y_train, y_test = train_test_split(features, target, test_size = 0.2)
```

Using grid search 

```python
grid_params = {'C': [0.1, 1, 10, 100, 1000],
              'gamma': [1, 0.1, 0.01, 0.001, 0.0001],
              'kernel': ['rbf', 'sigmoid']}
grid = GridSearchCV(SVC(),grid_params,refit=True, verbose=3)
grid.fit(x_train,y_train)
```

Start predicting

```python
predict = grid.predict(x_test)
report = classification_report(y_test,predict)
```
