Import libraries

```python
import pandas as pd
import numpy as np
import geopandas as gpd
from shapely.geometry import Point
import matplotlib.pyplot as plt 
import folium
import plotly_express as px
import json
import seaborn as sns; sns.set()  # for plot styling
```

Load data by districts

```python
data = gpd.read_file('/content/CSL_HCMC/Data/GIS/Population/population_HCMC/population_shapefile/Population_District_Level.shp')
data.head()
```

Sort district descending by Pop2019/Pop2017

```python
sorted_district= data.assign(raising = lambda x: x['Pop_2019']/ x['Pop_2017']).sort_values(by='raising',ascending=False)
print(sorted_district)
```

Load click json file

```python
clickData= gpd.read_file('/content/response3.json')
```

Convert data to crs type 

```python
 clickData=clickData.to_crs(epsg=32648)
```

```python
top10_click=clickData[:10]
print(top10_click)
```

Spartial join clicks and districts

```python
spartial_join=gpd.sjoin(clickData, sorted_district,how='left',op='within')
print(spartial_join)
```

Sort top 10 district with raising again

```python
top10_district_after_join=spartial_join.sort_values(by='raising',ascending=False)
top10_district_after_join.head(10)
```
